```python
def draw_circle(radius = 1):
    fig, ax = plt.subplots(figsize=[5,5])
    ax.set_xlim(-(radius + .25), radius + 0.25)
    ax.set_ylim(-(radius + .25), radius + 0.25)
    circle1 = plt.Circle((0, 0), radius, edgecolor='b',facecolor='none')
    rect = patches.Rectangle((-radius,-radius),radius*2,radius*2,linewidth=2,edgecolor='r',facecolor='none')
    ax.add_patch(rect)
    ax.add_artist(circle1)
    return ax

draw_circle()
plt.show()
```

```python
# Assume origin : 0,0
def _is_insideCircle(_x,_y, radius):
    if ((_x*_x) + (_y*_y)) <= radius*radius :
        return True
    return False
```

```python
class sampler:
    def __init__(self,radius=0.5):
        with pm.Model() as model:
            self.x = pm.Uniform('x', lower=-radius, upper=radius)
        with pm.Model() as model:
            self.y = pm.Uniform('y', lower=-radius, upper=radius)
    
    def get_sample(self):
        _x = self.x.random()
        _y = self.y.random()
        return np.array([_x ,_y])
            
```

```python
radius=1
num_samples = 20000
sampler_obj = sampler(radius)
count_in = 0
count_total = 0
samples_in = []
samples_out = []
```

```python
for _ in range(num_samples):
    xy = sampler_obj.get_sample()
    if _is_insideCircle(xy[0],xy[1],radius):
        count_in +=1
        samples_in.append(xy) 
    else:
        samples_out.append(xy) 
```

```python
samples_in = np.array(samples_in)
samples_out = np.array(samples_out)
```

## Plot the points sampled 

```python
ax = draw_circle()
ax.plot(samples_in[:,0],samples_in[:,1],'m.',markersize=0.5,alpha=0.55)
ax.plot(samples_out[:,0],samples_out[:,1],'b.',markersize=0.5,alpha=0.35)    
plt.show()
```

```python
pi = 4*count_in / num_samples
```
